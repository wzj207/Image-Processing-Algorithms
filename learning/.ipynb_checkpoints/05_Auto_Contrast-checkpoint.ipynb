```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('../images/low_contrast_1.jpg', cv2.IMREAD_GRAYSCALE)

img_copy = img.copy()

height = img.shape[0]
width = img.shape[1]

# min = 255
# max = 0
# for i in np.arange(height):
#     for j in np.arange(width):
#         a = img.item(i,j)
#         if a > max:
#             max = a
#         if a < min:
#             min = a

min = img_copy.min()
max = img_copy.max()

for i in np.arange(height):
    for j in np.arange(width):
        a = img_copy.item(i,j)
        b = float(a - min) / (max - min) * 255
        img_copy.itemset((i,j), b)

cv2.imwrite('../images/auto_contrast_new.jpg', img_copy)


f,(ax0,ax1) = plt.subplots(1,2,figsize=(20,15))

ax0.imshow(img,'gray')
ax0.axis('off')
ax0.set_title('original')

ax1.imshow(img_copy,'gray')
ax1.axis('off')
ax1.set_title('result')

```
