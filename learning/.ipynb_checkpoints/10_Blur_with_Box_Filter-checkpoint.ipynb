```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('../images/cat.jpg',cv2.IMREAD_GRAYSCALE)
img_out = img.copy()

height = img.shape[0]
width = img.shape[1]

for i in np.arange(3,height-3):
    for j in np.arange(3,width-3):
        sum = 0
        for k in np.arange(-3,4):
            for l in np.arange(-3,4):
                a = img.item(i+k,j+l)
                sum += a
        b = int(sum / 49.0)
        img_out.itemset((i,j),b)
```

```python
f,(ax0,ax1) = plt.subplots(1,2,figsize=(20,18))

ax0.imshow(img,'gray')
ax0.set_title('Original')
ax0.axis('off')

ax1.imshow(img_out,'gray')
ax1.set_title('Result')
ax1.axis('off')
```

```python
cv2.imwrite('../images/filter_box_new.jpg',img_out)
```
