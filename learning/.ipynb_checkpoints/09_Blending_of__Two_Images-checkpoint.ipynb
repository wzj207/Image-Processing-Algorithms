```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('../images/cat.jpg',cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('../images/lion.jpg',cv2.IMREAD_GRAYSCALE)

f,(ax0,ax1) = plt.subplots(1,2,figsize=(20,18))

ax0.imshow(img1,'gray')
ax0.axis('off')
ax0.set_title('Cat')

ax1.imshow(img2,'gray')
ax1.axis('off')
ax1.set_title('Lion')
```

```python
(height,width) = img1.shape[0],img1.shape[1]
height,width
```

```python
alpha = 0.5

img_out = img1.copy()

for i in np.arange(height):
    for j in np.arange(width):
        a1 = img1.item(i,j)
        a2 = img2.item(i,j)
        b = alpha * a1 + (1 - alpha) * a2
        img_out.itemset((i,j),b)

cv2.imwrite('../images/linear_blend_new.jpg',img_out)

# cv2.imshow('result',img_out)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
plt.imshow(img_out,'gray')
plt.title('Result')
plt.axis('off')
```
