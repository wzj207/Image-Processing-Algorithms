```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../images/low_contrast_1.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img,'gray')
```

```python
print(type(img))
print(img.shape)
print('Values Min/Max:',img.min(),img.max())
```

```python
height = img.shape[0]
width = img.shape[1]
```

```python
brightness = 50
```

```python
for i in np.arange(height):
    for j in np.arange(width):
        a = img.item(i,j)
        b = a + brightness
        if b > 255:
            b = 255
        img.itemset((i,j),b)
```

```python
plt.imshow(img,'gray')
```

```python
cv2.imwrite('../images/brightness_new.jpg',img)
```
