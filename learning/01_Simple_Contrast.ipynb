```python
import numpy as np
from skimage import io
import cv2
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# img = io.imread('../images/low_contrast_1.jpg',as_grey=True)
img = cv2.imread('../images/low_contrast_1.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
height = img.shape[0]
width = img.shape[1]
height,width,img.min(),img.max()
```

```python
plt.imshow(img,cmap='gray')
```

```python
contract = 1.3

for i in np.arange(height):
    for j in np.arange(width):
        a = img.item(i,j)
        b = math.ceil(a * contract) # math.ceil(null) returns integer 0 and does not give a NaN error.
        if b >= 255:
            b = 255
        img.itemset((i,j),b)

cv2.imwrite('../images/contrast_new.jpg',img)
```
