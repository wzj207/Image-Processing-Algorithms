```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('../images/low_contrast_1.jpg',cv2.IMREAD_GRAYSCALE)

img_2 = img.copy()

height,width = img_2.shape[0],img_2.shape[1]

print(type(img_2))
print(height,width)
print(img_2.min(),img_2.max())
```

```python
threshold = 150
```

```python
for i in np.arange(height):
    for j in np.arange(width):
        a = img_2.item(i,j)
        if a > threshold:
            b = 255
        else:
            b = 0
        img_2.itemset((i,j),b)
```

```python
f,(ax0,ax1) = plt.subplots(1,2,figsize=(15,10))
ax0.imshow(img,cmap='gray')
ax0.axis('off')
ax0.set_title('Original Image')

ax1.imshow(img_2,cmap='gray')
ax1.axis('off')
ax1.set_title('Result Image')
```

```python
cv2.imwrite('../images/threshold_new.jpg',img_2)
```
