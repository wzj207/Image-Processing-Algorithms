```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def histogram(img):
    height = img.shape[0]
    width = img.shape[1]
    
    hist = np.zeros(256)
    
    for i in np.arange(height):
        for j in np.arange(width):
            a = img.item(i,j)
            hist[a] += 1
            
    return hist

def cumulative_histogram(hist):
    cum_hist = hist.copy()
    
    for i in np.arange(1,256):
        cum_hist[i] = cum_hist[i-1] + cum_hist[i]
    
    return cum_hist
```

```python
img = cv2.imread('../images/img1.jpg',cv2.IMREAD_GRAYSCALE)
img_ref = cv2.imread('../images/cat.jpg',cv2.IMREAD_GRAYSCALE)

imgs = [img,img_ref]
f,axes = plt.subplots(1,2,figsize=(20,15))

for i,(im,ax) in enumerate(zip(imgs,axes)):
    axes[i].imshow(imgs[i],cmap='gray')
    axes[i].axis('off')
```

```python
hist = histogram(img)
hist_ref = histogram(img_ref)

cum_hist = cumulative_histogram(hist)
cum_hist_ref = cumulative_histogram(hist_ref)
```
